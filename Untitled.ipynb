```python
import pandas as pd
import numpy as np
import os
path = 'C:/Users/Admin/Downloads/data/'
dataframe = pd.read_csv(path +'sales2019_2.csv')
dataframe.head(10)
```

```python

import matplotlib.pyplot as plt
```

```python
os.listdir(path)
```

```python
all_length = []
frames = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path +file
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)
    
        length_1month = len(df1.index) 
        all_length.append(length_1month)
        #print(all_length)
print(sum(all_length))
#result
```

```python
df = result
df.to_csv('annualSales2019.csv')
df.head(10)
```

```python
df["Month"]=''
df.head(5)
df['Month']
```

```python
df['Month']=df["Order Date"].str[0:2]
```

```python
print(set(df['Month']))
```

```python
df = df.dropna(how='any')
df = df[df['Month'] !='Or']
df.head()
```

```python
#Task 3: Reporting
#3.1: What was the best month for sales? How much was earned that month?
df ['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'],downcast='integer')
df ['Price Each'] = pd.to_numeric(df['Price Each'],downcast='integer')
df['Sale'] = df['Quantity Ordered']*df['Price Each']
df.head()
```

```python
moving_column = df.pop('Sale')
df.insert(4,'Sale',moving_column)
df.head()
```

```python
df.groupby('Month').sum()['Sale']
```

```python
sale_value = df.groupby('Month').sum()['Sale']
sale_value.max()
```

```python
pip install matplotlib
```

```python
month = range (1,13)
plt.bar(x = month, height = sale_value)
plt.xticks(month)
plt.xlabel('Month')
plt.ylabel('Sales in USD')
plt.show()
```

```python
#TAsk 3.2: What city has the best sales? 
df['City']=''

df['City'] = df['Purchase Address'].str.slice(start = -9,step= 1, stop=-6)
df.head()
```

```python
address_to_city= lambda address:address.split(',')[1]
df['City']= df['Purchase Address'].apply(address_to_city)
df.head()
```

```python
sale_value_city=df.groupby('City').sum()['Sale']
sale_value_city.max()
set(df['City'])
```

```python

city_name=[]
for city, sales in sale_value_city.items():
    city_name.append(city)
print(city_name)
plt.bar(x = city_name , height = sale_value_city)
plt.xticks(city_name, rotation = 90)
plt.xlabel('City Name')
plt.ylabel('Sales in USD')
plt.show()
```

```python
#Why do you think it sold the most? =>> price,branch, quality

prices = df.groupby('Product').mean()['Price Each']
prices
```

```python
x = product_ls
y1 = quantity_value_product
y2 = prices
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x,y1,color = 'g')
ax2.plot(x,y2,color = 'b')

ax1.set_xticklabels(product_ls,rotation=90,size = 8)
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered', color = 'g')
ax2.set_ylabel('Price Each', color = 'b')
plt.show()
```
